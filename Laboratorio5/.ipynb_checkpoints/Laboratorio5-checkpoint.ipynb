## Boostrap o remuestreo

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import scipy 
from IPython.display import display, Math
import ipywidgets as widgets
import random
import scipy.stats as stats
import statistics
```

### Ejercicio 1
En un estudio sobre plantaciones de coihue, se ha recopilado información sobre la altura que alcanzan dichos árboles a los 20 años de edad, en diversas zonas de la región sur austral de Chile. A partir de dichos datos, estamos interesados en calcular un intervalo de confianza para la mediana de altura que alcanza esta especie a los 20 años de edad.
Para ello utilizaremos remuestreo o bootstrap.
1. Analice el código y evalue el número de muestras de bootstrap para alcanzar convergencia de la varianza estimada de la mediana.
2. Calcule el error cuadrático medio de la mediana.
3. Escoja otro estadístico de la muestra y calcule su incerteza con la técnica de boostrap
4. Compare y comente sus resultados.

```python
datos = [12.5,11.4,18.4,10.7,9.5,11.7,15.4,12.3,18.6, 14.7, 10.7,11.6, 10.1, 8.2, 11.4,21.1, 12.1, 7.5, 12.5, 14.7, 10.3, 5.4, 9.4, 13.8, 9.6, 11.3, 14.8, 12.6]
len(datos)
```

```python
n = len(datos)
m = statistics.median(datos)
print (m)
b=100
mm = 0
mn =[]
for i in range(b):
    dat = np.random.choice(list(datos),size=n,replace=True)
    mn.append(statistics.median(dat))
    
var_real = statistics.variance(mn)
print(var_real)
rango = 1.96*math.sqrt(var_real)
display(Math('%0.2f \pm %0.2f' % (m, rango )))

              
```

### Respuesta

### 1.1)

```python
tol = 1e-3
n = len(datos)
b = 90
idx = 0
dif = 1
conv = False
var_b = []
while(not conv):
    b+=10
    mn =[]
    for i in range(b):
        dat = np.random.choice(list(datos),size=n,replace=True)
        mn.append(statistics.median(dat))
        
    var = statistics.variance(mn)
    #print(var)
    var_b.append(var)
    if(idx > 1):
        if((abs(var_b[idx] - var_b[idx-1]) < tol) and (abs(var_b[idx-1] - var_b[idx-2]) < tol)):
            conv = True
    idx+=1    
    
print("La varianza estimada de la mediana converge con un valor de {}".format(var))
print("Se ha alcanzado la convergencia con b = ",b)   
```

Luego de experimentar con el valor de p, se obtuvo que aproximadamente con un valor de 2630 se logra una convergencia en cuanto a la varianza de la mediana.

```python
fig, ax = plt.subplots(figsize=(12,8))
x = np.arange(100,b+10,10)
ax.plot(x,var_b, label="Varianza estimada de la mediana")
ax.axhline(y=var, color='r', linestyle='-', label='Punto de convergencia estimado de la mediana')

ax.set_xlabel("Cantidad de muestras")
ax.set_ylabel("Varianza estimada de la mediana")
ax.set_title("Gráfica de convergencia")
ax.legend()
```

Se observa que aproximadamente con p = 2630, es decir, 2630 muestras de bootstrap se logra una convergencia de la varianza estimada de la mediana considerando una tolerancia igual a 1e-3.

### 1.3)

### Estimación de la incerteza para la curtosis.

Para el análisis se eligió la curtusis como estadístico, con un tamaño de b similar al calculado en el inciso anterior, ya que se considera una cantidad de muestras aceptable.

```python
b = 2700
kurt = stats.kurtosis(datos)
n = len(datos)
print (kurt)

kurt_list =[]
for i in range(b):
    dat = np.random.choice(list(datos),size=n,replace=True)
    kurt_list.append(stats.kurtosis(dat))
    
var = statistics.variance(kurt_list)
print(var)
rango = 1.96*math.sqrt(var)
display(Math('%0.2f \pm %0.2f' % (kurt, rango )))
```

En base a las pruebas de bootstrap realizadas, se estimó el nivel de incerteza de la curtosis, el cual toma valores en el rango 0.64  ±  1.90, donde puede concluirse que hay un 95% de confianza de que el valor de la curtosis se encuentre en este rango.

### Ejercicio 2

Suponga que los datos previamente registrados provienen de una distribución normal de parámetros $(\mu,\sigma)$

1. Calcule los estimadores máximo verosímil de $\hat{\mu}$ y $\hat{\sigma}$
2. Utilice boostrap paramétrico para estimar la incerteza de $\hat{\mu}$ y $\hat{\sigma}$ . Estime el número de muestras para alcanzar convergencia en cada caso.
3. Compare y comente sus resultados.



### Respuesta

### 2.1)

Para el cálculo de los estimadores máximo verosímiles, ya que se sabe que los datos pertenecen a una distribución normal se utiliza la función stats.norm.fit().

```python
mu, sigma = stats.norm.fit(datos)

print("Media mu: {}\nDesviación estandar sigma: {}".format(mu, sigma))

```

Mediante la estimación de máxima verosimilitud obtenemos que $\hat{\mu}$ tiene un valor de 12.225 y $\hat{\sigma}$ tiene un valor de 3.34.

### 2.2)

### Estimación de la incerteza de $\hat{\mu}$ con bootstrap paramétrico.

```python
tol = 1e-3
b = 90
idx = 0
dif = 1
var_b = []
conv = False
while(not conv):
    b+=10
    mean_list =[]
    for i in range(b):
        dat = np.random.choice(list(datos),size=n,replace=True)
        mean_list.append(statistics.mean(dat))
        
    var = statistics.variance(mean_list)
    var_b.append(var)
   
    if(idx > 1):
        if((abs(var_b[idx] - var_b[idx-1]) < tol) and (abs(var_b[idx-1] - var_b[idx-2]) < tol)):
            conv = True
    idx+=1    
print("Se ha alcanzado la convergencia con b = ",b) 
```

```python
fig, ax = plt.subplots(figsize=(12,8))
x = np.arange(100,b+10,10)
ax.plot(x, var_b, label="Varianza estimada de la media")
ax.axhline(y=var, color='r', linestyle='-', label='punto de convergencia estimado')
ax.set_xlabel("Cantidad de muestras")
ax.set_ylabel("Varianza estimada de la media")
ax.set_title("Gráfica de convergencia")
ax.legend()
```

```python
var_real = statistics.variance(mean_list)
print(var_real)
rango = 1.96*math.sqrt(var_real)
display(Math('%0.2f \pm %0.2f' % (mu, rango )))
```

En base a las pruebas de bootstrap realizadas, se estimó el nivel de incerteza del estadístico $\mu$, el cual toma valores en el rango 12.23 $\pm$ 1.23, donde puede concluirse que hay un 95% de confianza de que el valor de $\mu$ se encuentre en este rango.
